```python
import pandas as pd
```

```python
gpa_act_data = pd.read_csv('CH01PR19.txt', delim_whitespace=True, header=None)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
X = gpa_act_data[[1]]
y = gpa_act_data[[0]]
```

```python
linreg_obj = LinearRegression()
linreg_obj.fit(X, y)
```

```python
beta_1 = linreg_obj.coef_
print("Beta 1:", beta_1)
beta_0 = linreg_obj.intercept_
print("Beta 0:", beta_0)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(X, y)
plt.plot(X, linreg_obj.predict(X), color='red')
plt.xlabel('ACT Score')
plt.ylabel('GPA')
plt.title('ACT Score vs GPA with Linear Regression Line')
plt.show()
```

```python
import numpy as np 
linreg_obj.predict(np.array([[30]]))
```

```python
beta_1 = linreg_obj.coef_
print("Point estimate of the change in the mean response when the entrance score increases by one point:", beta_1)
```

```python
# question 2
crime_rate_data = pd.read_csv('CH01PR28.txt', delim_whitespace=True, header=None)
```

```python
X = crime_rate_data[[1]]
y = crime_rate_data[[0]]
```

```python
linreg_obj = LinearRegression()
linreg_obj.fit(X, y)
```

```python
beta_1 = linreg_obj.coef_
print("Beta 1:", beta_1)
beta_0 = linreg_obj.intercept_
print("Beta 0:", beta_0)
```

```python
plt.scatter(X, y)
plt.plot(X, linreg_obj.predict(X), color='red')
plt.xlabel('High School Graduation Rate')
plt.ylabel('Crime')
plt.title('High School Graduation Rate vs Crime with Linear Regression Line')
plt.show()
```

```python
# This is simply the coefficient (slope) of the regression line
slope = linreg_obj.coef_
print(f"1. Change in crime rate for 1% increase in graduation rate: {slope}")

# 2. Mean crime rate for X = 80
X_new = np.array([[80]])  # Need 2D array for predict
predicted_crime_rate = linreg_obj.predict(X_new)[0]
print(f"2. Predicted crime rate at 80% graduation rate: {predicted_crime_rate}")

# 3. Calculate ε₁₀ (residual for 10th observation)
y_pred = linreg_obj.predict(X)
epsilon_10 = y.loc[9] - y_pred[9]  # Using 0-based indexing
print(f"3. ε₁₀ (residual for 10th observation): {float(epsilon_10)}")

# 4. Calculate σ² (variance of residuals)
residuals = y - y_pred
sigma_squared = np.sum(residuals**2) / (len(y) - 2)  # n-2 degrees of freedom
print(f"4. σ² (variance of residuals): {float(sigma_squared)}")
```
